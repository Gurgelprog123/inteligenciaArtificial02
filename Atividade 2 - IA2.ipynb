```python
import os
import numpy as np
import random as rnd
```

```python
"""
Pseudo código da codificação binária:

Rotina para criação de população inicial aleatória
Nind: número de indivíduos da população
Ncrom: número de cromossomos (parâmetros) por indivíduo
CromLim: matriz (Ncrom x 2) contendo limites superior e inferior para cromossomos (parâmetros)
"""

def newpop(Nind: int, CromLim: np.array):
    Ncrom = CromLim.shape[0]  # número de cromossomos
    pop = np.zeros((Nind, Ncrom))  # cria matriz da população

    for i in range(Nind):
        for j in range(Ncrom):
            inf = CromLim[j, 0]
            sup = CromLim[j, 1]
            pop[i, j] = (rnd.random() * (sup - inf) + inf)
    return pop
```

```python
"""
Pseudo código da codificação binária:

Rotina para codificação binária de indivíduos
pop: população a ser codificada
Ncrom: número de cromossomos em cada indivíduo
CromLim: matriz Ncrom x 2 contendo os limites inferior e superior para os cromossomos
Lbits: vetor Ncrom contendo o numero de bits de cada cromossomo
"""

def code(pop: type(newpop), CromLim: np.array, Lbits: np.array):
  Npop = pop.shape[0]
  NcromLim = CromLim.shape[0]
  cod = np.array([[], []])

  for i in range(Npop):
      for j in range(NcromLim):
          inf = CromLim[j, 0]
          sup = CromLim[j, 1]
          valor_real = pop[i, j]
          aux = int((valor_real-inf)/(sup-inf)*(2**Lbits[j]-1))
          if j==1:
            temp=aux
          else:
            temp=str(aux)
      cod[i]=temp
      return cod
```
